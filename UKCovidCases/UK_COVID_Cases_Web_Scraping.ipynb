# Web-scraping COVID data
I want to analyse and visualise individual local authority covid-case data. However, the data is published by GOV.uk only by collection with tables for collections such as England, UK total, or the individual local authorities. I would like to scrape each individual local authority, and combine them so that I can plot each authority in a time series. 

Data sources: [UK Area boundaries](https://geoportal.statistics.gov.uk/datasets/ons::local-authority-districts-may-2021-uk-bfe/about), [COVID Data](https://coronavirus.data.gov.uk/details/cases?areaType=ltla&areaName=Aberdeen%20City)

References: [Medium article](https://towardsdatascience.com/how-to-use-selenium-to-web-scrape-with-example-80f9b23a843a)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.ops import unary_union

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Districts shapes and names from https://geoportal.statistics.gov.uk/datasets/ons::local-authority-districts-may-2021-uk-bfe/about

fp = '/datasets/Local_Authority_Districts/LAD_MAY_2021_UK_BFE_V2.shp'

df_dist = gpd.read_file(fp).set_index('LAD21NM').sort_index()

df_dist.head()

# index names here match Local authority names I'm scraping. I also now have the geography
```

```python
# Only really want area name (as index) and geometry

df_dist = gpd.GeoDataFrame(df_dist.geometry.rename_axis('local_auth'))

local_auths = list(df_dist.index)
```

```python
local_auths
```

## Snags & Preprocessing
- Local authority list lists Cornwall and Isles of Scilly as two separate areas, but are collected together for COVID data - need to ammend
- Similarly Hackney and City of London are combined - need to amend for scraping but also to amend geography
- Similarly North and West Northamptonshire are combined to Northampton

```python
# Let's remove combined features and add the correct label for scraping

areas_to_clean = ['City of London','Cornwall','Hackney','Isles of Scilly','North Northamptonshire','West Northamptonshire']

for area in areas_to_clean:
    
    local_auths.remove(area)

comb_areas = ['Hackney and City of London', 'Northampton', 'Cornwall and Isles of Scilly']

local_auths += comb_areas
```

## Web-scraping

We shoule be able to access each individual webpage now by looping through the local authority list, and using it to update the URL as the first URL is 'https://coronavirus.data.gov.uk/details/cases?areaType=ltla&areaName=Aberdeen%20City' for Aberdeen City.

Let's focus on getting the correct data out for Aberdeen City. Looking at the data table, we can see that the data we want sits in a 'date' and 'numeric' td type, inside of a table row, inside of a table body. A row is shown here:


    <tr class="sc-cNKqjZ PEdbg govuk-table__row  ">

        <td type="numeric" class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--numeric ">6</td>
        <td type="numeric" class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--numeric ">25,502</td>
    </tr>
    
We can use seleniums find_element_by_xpath methods to access the correct table and data.

```python
driver = webdriver.Chrome('/Users/gabrielcoulter/Downloads/chromedriver')

# Had some permission issues with macOS here

driver.get('https://coronavirus.data.gov.uk/details/cases?areaType=ltla&areaName=Aberdeen%20City')

# First need to click 'Data' Button - occasionaly this timedout so we can force it to wait until the button is clickable

data_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="card-cases_by_specimen_date"]/div[1]/div[1]/button[3]')))

data_button.click()


# Once clicked, we need to wait for the data table to load before scraping and moving on

table_xpath = '//*[@id="card-cases_by_specimen_date"]/div[1]/div[2]/div/table/tbody'#'//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--date "]'

element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, table_xpath)))

# Once the table element is detected, we can scrape the data

print('Retrieving data...')

dates_raw = driver.find_elements_by_xpath('//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--date "]')
daily_cases_raw = driver.find_elements_by_xpath('//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--numeric "]')[0:2*len(dates_raw):2]
# This second xpath pulls both daily cases and cumulative cases for each date
# Therefore we need every second element, hence the list indexing [0::2]

# We need to extract the data text from the retrieved selenium elements.
# TODO this is a huge bottleneck and needs to be done faster

print('Parsing data...')

dates = [date.text for date in dates_raw]
daily_cases = [case.text for case in daily_cases_raw]
    
print('Dates: ' + str(dates[:5]),'\nCase Numbers: ' + str(daily_cases[:5]))

driver.close()
```

That looks pretty good. All that's left now is to loop through the various local authorities, adding a column for each.

Looking at the URL, https://coronavirus.data.gov.uk/details/cases?areaType=ltla&areaName=Aberdeen%20City, it seems we can get each individual page by updating the last part of the URL with a formatted local authority

Aberdeen City - Aberdeen%20City

```python
# This list will contain names where there was an issue with the scraping process for troubleshooting

problem_regions = []
```

```python
# df = pd.DataFrame() # Uncomment this if it's the first time running the script - need a fresh dataframe for the data

driver = webdriver.Chrome('/Users/gabrielcoulter/Downloads/chromedriver')

for local_authority in local_auths[10:]:
    
    # The URL is slightly different for a handful of areas - these nested try-except blocks try both, and if there is
    # still an issue, puts the area name in a list for troubleshooting without stopping the scrape.
    
    try:
        url = 'https://coronavirus.data.gov.uk/details/cases?areaType=ltla&areaName=' + local_authority.replace(' ', '%20')
    
        driver.get(url)

        # First need to click 'Data' Button - occasionaly this timedout so we can force it to wait until the button is clickable

        data_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="card-cases_by_specimen_date"]/div[1]/div[1]/button[3]')))
        
    except:
        
        try:
            url = 'https://coronavirus.data.gov.uk/details/cases?areaType=utla&areaName=' + local_authority.replace(' ', '%20')

            # most are lower local auths, some are upper replacing 'ltla' with 'utla'

            driver.get(url)

            # First need to click 'Data' Button - occasionaly this timedout so we can force it to wait until the button is clickable

            data_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="card-cases_by_specimen_date"]/div[1]/div[1]/button[3]')))
        
        except:
            
            problem_regions.append(local_authority)
            continue
            

    data_button.click()
    

    # Need to wait for the data to load before scraping and moving on
    
    table_xpath = '//*[@id="card-cases_by_specimen_date"]/div[1]/div[2]/div/table/tbody'#'//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--date "]'

    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, table_xpath)))
    
    
    # Now we can scrape the data
    
    dates_raw = driver.find_elements_by_xpath('//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--date "]')
    daily_cases_raw = driver.find_elements_by_xpath('//td[@class="sc-bBHHxi ineTJY govuk-table__cell govuk-table__cell--numeric "]')[0:2*len(dates_raw):2]
    # This pulls both daily cases and cumulative cases, therefore we need every second element, hence the list indexing. 

    
    dates = [date.text for date in dates_raw]
    daily_cases = [case.text for case in daily_cases_raw]
        
    df[local_authority] = pd.Series(data = daily_cases, index = pd.to_datetime(dates, dayfirst=True)) # Dayfirst super important here
    
driver.close()

df.head()
```

## Geometry cleaning

```python
df_dist.head()
```

```python
# The web-scraped data is correct, but we need to combine area polygons now for plotting

df_dist.loc[areas_to_clean, 'geometry']
```

```python
for area1, area2 in zip(['Hackney', 'North Northamptonshire', 'Cornwall'], ['City of London', 'West Northamptonshire', 'Isles of Scilly']):
    
    # Get correct combined area name
    
    comb_area = area1 + ' and ' + area2
    
    if area1 == 'North Northamptonshire':
        comb_area = 'Northampton'
        
    # Add combined geometry as a new entry and drop separate areas
    
    df_dist.loc[comb_area, 'geometry'] = unary_union([df_dist.loc[area1, 'geometry'], df_dist.loc[area2, 'geometry']])
    
    df_dist.drop([area1, area2], inplace=True)
```

## Plotting
In order to plot this data as a time-series, we need to combine the COVID data with the geometry data. This, unfortunately, requires us to make a wide-format dataframe; each local authority will have a row, and each date will be a column

```python
df.index
```

```python
dates = df.index.sort_values()
dates
```

```python
df_pivot = df.T

merged = df_dist.join(df_pivot).drop('num', axis=1)

merged.head()
```

```python
merged.columns = merged.columns[:1].tolist() + pd.to_datetime(merged.columns[1:]).tolist()
```

```python
merged.columns
```

```python
%matplotlib notebook
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

vmax, vmin = (max(df.max()), min(df.min()))

cmap_name = 'plasma'

merged.plot(column=pd.to_datetime(dates[-1]), cmap=cmap_name, ax=ax)

ax.axis('off')

ax.set_title('COVID Cases', fontdict={'fontsize':'16', 'fontweight':3}, alpha=0.5, ha='left',va='bottom', position=(0,1))

ax.annotate('Source: GOV.uk', xy=(0.125, 0.08), xycoords='figure fraction', ha='left', va='top', 
            fontsize=12, alpha=0.3)

sm = plt.cm.ScalarMappable(cmap=cmap_name, norm = plt.Normalize(vmin=vmin, vmax=vmax))

sm._A = []

cbar = fig.colorbar(sm, fraction=0.030, pad=0.04)
```
